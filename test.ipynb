```python
import yfinance as yf
import numpy as np
import pandas as pd
import datetime
from dateutil.relativedelta import relativedelta
```

```python
tickers = ["SPY", "VEU", "BIL", "AGG"]

start_date = datetime.datetime(2007, 6, 1)
end_date = datetime.datetime.now()


dat = yf.download(tickers, start=start_date, end=end_date, auto_adjust=True, interval="1d")

dat
```

```python
def precise_month_conversion(start_date, end_date):
    """
    Calculate months between two dates
    """
    delta = relativedelta(end_date, start_date)
    return delta.years * 12 + delta.months
```

```python
def calc_return(start, end):
    return end/start-1
```

```python
end = datetime.datetime(2025, 5, 2)
start = datetime.datetime(2007, 6, 1)

months = precise_month_conversion(start, end)
```

```python
allkeys = dat["Close"]["SPY"].index

selected_keys = list()

for month in range(months):
    currentdate = start + relativedelta(months=+month)
    datestring = pd.Timestamp(currentdate)

    if datestring in allkeys:
        selected_keys.append(datestring)
        print(datestring)
    else:
        print("Not included:", datestring)

        numdays = 0

        while datestring not in allkeys:
            numdays += 1

            currentdate = start + relativedelta(months=+month, days=-numdays)
            datestring = pd.Timestamp(currentdate)

        selected_keys.append(datestring)
        print("Replacement:", datestring)


print(selected_keys)
```

```python
calculated_data = dict()

for ticker in tickers:

    returns = []

    for i in range(len(selected_keys)-12):
        start = dat["Close"][ticker][selected_keys[i]]
        end = dat["Close"][ticker][selected_keys[i+12]]

        returns.append(calc_return(start, end))

    calculated_data[ticker] = returns

calculated_data["Date"] = selected_keys[12:]

df = pd.DataFrame(data=calculated_data)
df = df.set_index("Date")

df
```

```python
df1 = df[["SPY", "VEU", "BIL", "AGG"]]
df1["Date_next"] = df1.index.to_series().shift(-1)
df1 = df1.iloc[:-1 , :]

for index, row in df1.iterrows():
    a = np.max(row[:-2].array)
    break
df1
```

```python
idx_list = list()
ret_list = list()
ticker_list = list()
perf_list = list()

for index, row in df1.iterrows():
    ret = np.max(row[:-2].array)

    tick = row.index[np.argmax(row[:-2].array)]
    next_date = row["Date_next"]

    # print("Next:", next_date)
    # print("End:", dat["Close"][tick][next_date])

    if tick == "BIL":
        perf = calc_return(dat["Close"]["AGG"][index], dat["Close"]["AGG"][next_date])
    else:
        perf = calc_return(dat["Close"][tick][index], dat["Close"][tick][next_date])
    

    print(index, ret, tick, perf)
    idx_list.append(index)
    ret_list.append(ret)
    ticker_list.append(tick)
    perf_list.append(perf)



gem = pd.DataFrame({"Date": idx_list, "Max Return": ret_list, "Ticker": ticker_list, "Performance": perf_list})
gem = gem.set_index("Date")
gem
```

```python
def yearly_perf(start_date):
    begin = gem.index.get_loc(start_date)
    end = begin + 12

    data_array = gem["Performance"][begin:end].array
    data_array += 1
    return np.cumulative_prod(data_array)[-1]-1
```
